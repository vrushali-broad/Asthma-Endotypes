```python
import pandas as pd
import os
import datetime
from datetime import timedelta, date
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import scipy.stats as stats
from scipy.stats import ranksums, mannwhitneyu
import statsmodels.api as sm
```

#### <u> <font color=blue> Functions </font>  </u>

```python
def patients(df):
    return list(set(df['Biobank_Subject_ID']))

def common_patients(df1, df2):
    return list(set(df1['Biobank_Subject_ID']) & set(df2['Biobank_Subject_ID']))

def get_subset(df, col, val, neg = False):
    if neg:
        return df[df[col] != val]
    else:
        return df[df[col]==val]
```

```python
def boxplot_cells(df, cell, col_name, order, ylabel, ylim, width, title, name, pal = False):
    
    types = order
    fig, ax = plt.subplots(nrows=1,  figsize=(5.5,5.5)) 
    
    ## Properties
    medianprops = {'color': 'magenta', 'linewidth': 2}
    boxprops = {'edgecolor': 'k', 'linestyle': '-'}
    flierprops = dict(color='white',marker='o', markersize=1,markerfacecolor='auto', markeredgecolor='k')
    whiskerprops=dict(color='k',linewidth=1,linestyle='dashed')
    capprops=dict(color='black',linestyle= '-',linewidth=1)
    
    ## Plot
    if pal:
        palette = ['#84b800','#4c4cff','#ff4c93','#b75454']
    else:
        palette = None
    sns.boxplot(x= col_name, y = cell, data=df, 
                order = order, 
                flierprops=flierprops,
                capprops = capprops,# notch=True,
                whiskerprops = whiskerprops,
                boxprops = boxprops,
                ax=ax, 
                palette= palette,
                width= width)
    sns.pointplot(x=col_name, y=cell, data=df, order = order, ci=None, color='yellow',ax=ax, join=False)

    for i in range(len(types)-1):
        pval = stats.ranksums(df[df[col_name]==types[i]][cell], df[df[col_name]==types[i+1]][cell])
        x1, x2 = i+0.05, i+1-0.05 
        if cell == 'AVG_NEUT' :
            y, h, col = df[cell].mean() + 36 , 1.5, 'k'    
        elif cell == 'AVG_EOS':
            y, h, col = df[cell].mean() + 6 , 0.3, 'k'
        elif cell == 'AVG_BASO':
            y, h, col = df[cell].mean() + 0.8 , 0.03, 'k'
        if (pval.pvalue < 0.0005):
            ax.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
            ax.text((x1+x2)*.5, y+h,"***", ha='center', va='bottom', color=col, fontsize=18)
        elif (pval.pvalue < 0.005):
            ax.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
            ax.text((x1+x2)*.5, y+h,"**", ha='center', va='bottom', color=col, fontsize=18)
        elif (pval.pvalue < 0.05):
            ax.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
            ax.text((x1+x2)*.5, y+h,"*", ha='center', va='bottom', color=col, fontsize=18)
        print(types[i],types[i+1], pval.pvalue)
        ax.set_ylim(ylim)
        ax.set_ylabel('')    
        ax.set_xlabel('')
    ax.tick_params(axis='both',  labelsize=14)
    ax.axhline(np.mean(df[cell]), ls='--', color='cyan')
    order = [line.replace(' ', '\n') for line in order]
    ax.set_xticklabels(order)#,  rotation = 90)
    ax.set_title(title, fontsize= 14, fontweight='bold') 
    ax.set_ylabel(ylabel, fontsize= 14, fontweight='bold')
    plt.savefig(name, bbox_inches = "tight")
    # plt.savefig('Clin_neutro.svg',bbox_inches = "tight")
```

```python
def autolabel(rects, color=False):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2. , 1.001*height,
                str('%.2f' % height)+'%',fontsize = 10, c = color,
                ha='center', va='bottom')
        
def autolabel_no_pert(rects, color=False):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2. , 1.001*height,
                str('%.2f' % height),fontsize = 10, c = color,
                ha='center', va='bottom')

def barplot_pert(bars, pert, width, title, ylim, name, ylabel, deceased= None, no_pert = False):
    plt.figure(figsize = [5, 5])        
    y_pos = np.arange(len(bars))

    barplot_endotypes = plt.bar(y_pos, pert, color='grey', width=width, edgecolor='white', label='COPD')
    if no_pert:
        autolabel_no_pert(barplot_endotypes, color = 'k')
    else:
        autolabel(barplot_endotypes, color = 'k')
    fontsize = 14
    plt.title(title, y=1.001, fontsize= fontsize,fontweight='bold')
    if ylabel:
        plt.ylabel(ylabel, fontweight = 'bold', fontsize = 12)

    plt.yticks(fontsize=fontsize)
    plt.ylim(ylim)
    bars = [line.replace(' ', '\n') for line in bars]
    plt.xticks([r for r in range(len(bars))], bars, fontsize=12)
    plt.savefig(name, bbox_inches = "tight")
    
def barplot_overlay(bars, pert1, pert2, width, title, ylim, name, labels, copd = None):
    plt.figure(figsize = [5, 5])        
    y_pos = np.arange(len(bars))

    barplot_endotypes = plt.bar(y_pos, pert1, color='grey', width=width, edgecolor='white', label=labels[0])
    autolabel(barplot_endotypes, color = 'k')
    
    barplot_endotypes = plt.bar(y_pos, pert2, color='silver', width=width, edgecolor='white', label=labels[1])
    autolabel(barplot_endotypes, color = 'aqua')
    
    fontsize = 14
    plt.title(title, y=1.001, fontsize = fontsize, fontweight='bold')
    if copd:
        plt.ylabel('% Patients with COPD', fontweight='bold', fontsize=12)
    else:
        plt.ylabel('% Deceased Patients', fontweight='bold', fontsize=12)
    
    plt.yticks(fontsize=fontsize)
    plt.ylim(ylim)
    plt.legend()
    bars = [line.replace(' ', '\n') for line in bars]
    plt.xticks([r for r in range(len(bars))], bars, fontsize=12)
    plt.savefig(name, bbox_inches = "tight")
    
```

```python
def trend_analysis(df, cond, group):
    print(cond)
    data = df[[cond, group]]
    table = sm.stats.Table.from_data(data) 
    if group == 'Endotype':
        table.table_orig = table.table_orig.loc[:,endo_order]
    else:
        table.table_orig = table.table_orig.loc[:,pheno_order]
    data = np.array([table.table_orig.loc['Yes'], table.table_orig.loc['No']])
    table = sm.stats.Table(data)
    print(table)
    rslt = table.test_ordinal_association()
    print(rslt.pvalue)
    print('****************')
```
